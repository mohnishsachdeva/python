```python
def check_pearson(column_name):
    r = pearsonr(selected_dataset['diagnosis'],selected_dataset[column_name])
    return abs(r[0])
```

```python
def type_of_column(column_name):
    if type(selected_dataset[column_name].iloc[0]) == str or type(selected_dataset[column_name].iloc[0])==int and len(selected_dataset[column_name].unique()) < len(selected_dataset[column_name]):
        if len(selected_dataset[column_name].unique()) == 2:
            return 'binomial'
        else:
            return 'multinomial'
    else:
        est_normal = normal_likelihood_function(column_name)
        
        est_rayleigh = rayleigh_likelihood_function(column_name)
        
        if est_normal > est_rayleigh:
            return 'normal'
        else:
            return 'rayleigh'
        
```

\begin{equation}
\mu_{\text{cap}}=\frac{\sum\limits_{i=1}^n{x_i}}{N}
\end{equation}


```python
def normal_likelihood_function(column_name):
    mu_cap = selected_dataset[column_name].mean()
    sigma_cap = selected_dataset[column_name].std()
    estimate_normal = np.sum(s.norm.logpdf(selected_dataset[column_name],loc=mu_cap,scale=sigma_cap))
    return estimate_normal
```

\begin{equation}
\sigma_{\text{cap}} = \sqrt{\frac{\sum\limits_{i=1}^n({x_i}-{\mu}_{cap})^{2}}{N}}
\end{equation}

```python
def rayleigh_likelihood_function(column_name):
    sigma_cap = np.sqrt(np.mean(selected_dataset[column_name].apply(lambda x: x**2))/2)
    estimate_rayleigh = np.sum(s.rayleigh.logpdf(selected_dataset[column_name],scale=sigma_cap))
    return estimate_rayleigh
```

\begin{equation}
\sigma_{\text{cap_rayleigh}}= \sqrt{\frac{\sum\limits_{i=0}^{n}{x_i}^2}{2n}}
\end{equation}

\begin{equation}
\text{Naive Bayes Classifier}\\
P(A|B) = \frac{P(B|A)*P(A)}{P(B)}
\end{equation}

\begin{equation}
\text{P(diagnosis = 'B'| concave point_worst = x)} = \frac{\text{P(concave point_worst = x | diagnosis = 'B') * P(diagnosis = 'B')}}{\text{P(concave point_worst = x)}}
\end{equation}

\begin{equation}
\text{Posterior Probability}\\\text{P(A|B)= P(Diagnosis='B'|concave point_worst = x)}
\end{equation}

\begin{equation}
\text{Likelihood probability}\\\text{P(A|B) = P(concave point_worst = x | diagnosis = 'B')}
\end{equation}

\begin{equation}
\text{Prior Probability}\\ \text{P(A)= P(Diagnosis='B')}
\end{equation}

\begin{equation}
\text{Evidence or Normalizing Probability }\\ \text{P(B) = P(concave point_worst = x)}
\end{equation}
