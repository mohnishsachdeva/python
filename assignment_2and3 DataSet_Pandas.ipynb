```python
column_names=new_df.columns
column_names
```

\begin{equation}
\mu_{cap}=\frac{\sum_{i=1}^n{x_i}}{N}
\end{equation}


```python
def best_estimates(column_name_func):
    mu_cap = new_df[column_name_func].mean()
    sigma_cap = new_df[column_name_func].std()
    mode_cap = np.sqrt(np.mean(new_df[column_name_func]**2)/2)
    return [column_name_func,mu_cap,sigma_cap,mode_cap] # column_name,mucap,sigma_cap,mode_cap
```

```python
best_estimates_value = []
for column in column_names:
    best_estimates_value.append(best_estimates(column))
```

```python
lf_normal = []
for i in range(0,len(best_estimates_value)):
    lf_normal.append(np.prod(s.norm.pdf(new_df[best_estimates_value[i][0]],best_estimates_value[0][1],best_estimates_value[0][2])))
lf_normal
```

```python
lf_rayleigh = np.prod(s.rayleigh.pdf(new_df[best_estimates_value[0][0]],best_estimates_value[0][1],best_estimates_value[0][3]))
lf_rayleigh
```

```python
def check_distribution(lf_normal_func,lf_rayleigh_func):
    if lf_normal_func > lf_rayleigh_func:
        return 'symmetric'
    else:
        return 'right skewed'
```

```python
output[best_estimates_value[0][0]]= check_distribution(lf_normal,lf_rayleigh)
```
